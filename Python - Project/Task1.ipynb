```python
import pandas as pd
import numpy as np
ae = pd.read_csv("FEV-data-Excel.xlsx - Auta elektryczne.csv")   #Read the csv file using Pandas library
null_values = ae.isnull().sum()               #Find all the NAN/Null values 
print(null_values)
print(ae.info())
ae['Type of brakes'].fillna("disc (front + rear)", inplace=True)    #Replacing NAN value of "Type of Brakes" column to a certain value
ae.fillna(ae.median(numeric_only=True).round(1),inplace = True)              #Replacing NAN of numerical only column to Median values
pd_ae = pd.DataFrame(ae)
display(pd_ae)
print(ae.info())
ae.to_csv('cleaned_FEV-data-Excel.xlsx - Auta elektryczne.csv', index=False)  #Renaming and Saving the cleaned file
```
