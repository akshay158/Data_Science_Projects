```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
ae = pd.read_csv("cleaned_FEV-data-Excel.xlsx - Auta elektryczne.csv")
```

```python
# Plot Engine Power vs. Price to identify pricing trends related to engine performance
plt.figure(figsize=(10, 6))
sns.lineplot(data=ae, x='Engine power [KM]', y='Minimal price (gross) [PLN]', ci=None)
plt.title("Trend: Engine Power vs. Price")
plt.xlabel("Engine Power [KM]")
plt.ylabel("Minimal Price (gross) [PLN]")
plt.show()

# As per line chart, it shows positive trend (line going upwards), which means as the Engine Power of EVs increases, it's Gross Price will also
# will also increase upto a certain point then it will come down a little as it may become too expensive for customers to buy.
```

```python
# Plot Weight vs. Acceleration to identify trends related to weight vs Acceleration of vehicle
plt.figure(figsize=(10, 6))
sns.lineplot(data=ae, x='Minimal empty weight [kg]', y='Acceleration 0-100 kph [s]', ci=None)
plt.title("Trend: Weight vs Acceleration")
plt.xlabel("Minimal empty weight [kg]")
plt.ylabel("Acceleration 0-100 kph")
plt.show()

#As per line chart, it can be said that it shows slighlty negative trend, we can say that as the weight of Vehicle will increases, 
#the acceleration time of vehicle, to go from 0-100 will increase. The weight of vehicle have negative impact on Acceleration of those vehicle.
# Too much addition of weight to the vehicle must be considered as it will affect vehicle's acceleration and thus not satisfy some customers.
```
