```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
ae = pd.read_csv("cleaned_FEV-data-Excel.xlsx - Auta elektryczne.csv")
manufacturer = ae.groupby('Make').agg({
    "Minimal price (gross) [PLN]":'mean',
    "Battery capacity [kWh]" : 'mean',
    "Range (WLTP) [km]":'mean',
    "Maximum speed [kph]":'mean',
    "mean - Energy consumption [kWh/100 km]":'mean'}).round(2).sort_values(by="Range (WLTP) [km]", ascending=False)
print("Average metrics by manufacturer: ")
print(manufacturer)
```

```python
#To determine performance of all the Make who manufactures EV, we can analyze them accross various factors:
# Analyzing the Make as which manufacturer offers longer driving range and higher Battery Capacity in their vehicles, 
# we can plot the mean calculated for range and battery capacity amongst each other.

plt.figure(figsize=(12,8))
sns.barplot(data=manufacturer, x="Battery capacity [kWh]",y='Range (WLTP) [km]',hue=manufacturer.index, width=2)
plt.title("Manufacturer Comparison: Battery Capacity vs. Range")
plt.xlabel("Battery Capacity [kWh]")
plt.ylabel("Range (WLTP) [km]")
plt.legend(title="Manufacturer", bbox_to_anchor=(1.1, 1))
plt.show()

# As per the graph, it can be determined that Tesla offers highest driving range and larger battery capacity compared to other Make.
# While Smart manufactures vehicles with least driving range and has least battery capacity.
```

```python
#Plotting Price against driving range to determine which Make offers best value against driving distance.

plt.figure(figsize=(12,8))
sns.barplot(data=manufacturer, y="Minimal price (gross) [PLN]", x = 'Range (WLTP) [km]', hue=manufacturer.index, width=2)
plt.title("Manufacturer Comparison: Price vs. Range")
plt.xlabel("Minimal price (gross) [PLN]")
plt.ylabel("Range (WLTP) [km]")
plt.legend(title="Manufacturer", bbox_to_anchor=(1.1, 1))
plt.show()

# As per scatterplot, it can be determined that Porsche offers same driving range as by Volkswagen but the price of Porsche is much higher than Volkswagen.
# Make like Tesla, Audi and Jaguar falls under similar price range but Tesla offers highest driving range compared to other manufacturers
```

```python
#Plotting Maximum Speed offered by each Make against its Energy Consumption to determine which Make has higher speed and is Energy efficient.

plt.figure(figsize=(12,8))
sns.barplot(data=manufacturer, x="Maximum speed [kph]", y = "mean - Energy consumption [kWh/100 km]", hue=manufacturer.index, width=2)
plt.title("Manufacturer Comparison: Maximum Speed vs. Energy Consumption")
plt.xlabel("Maximum speed [kph]")
plt.ylabel("mean - Energy consumption [kWh/100 km]")
plt.legend(title="Manufacturer", bbox_to_anchor=(1.1, 1))
plt.show()

#As per plot, Makes like Mercedes-Benz, Audi and Porsche are not enery efficient even though they offer good amount of speed with Porsche having highest speed.
# But Tesla is a brand that is highly energy efficient and still has second highest speed to offer.
```
