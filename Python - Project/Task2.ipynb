```python
'''
The dataset provides detailed information about various electric vehicles (EVs), 
capturing technical specifications, performance metrics, and physical dimensions.

Technical Specifications includes:
Battery Capacity [kWh]: Indicating energy storage capabilities, critical for driving range.
Range (WLTP): Indicating vehicles long distance travel range and energy efficiency
Maximum DC Charging Power [kW]: Highlighting advancements in charging technology for reduced downtime.
Maximum Speed and Acceleration: Highlighting the top speed og vehicles and fast is their acceleration time.
'''
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
ae = pd.read_csv("cleaned_FEV-data-Excel.xlsx - Auta elektryczne.csv")
tech_spec_data = ae[ae['Battery capacity [kWh]'].notnull()]
plt.figure(figsize=(7, 5))
sns.scatterplot(data=tech_spec_data, x='Battery capacity [kWh]', y='Range (WLTP) [km]')
plt.title('Technical Specifications: Battery Capacity vs. Range')

plt.xlabel('Battery Capacity [kWh]')
plt.ylabel('Range (WLTP) [km]')
plt.grid(True)
plt.show()


```
