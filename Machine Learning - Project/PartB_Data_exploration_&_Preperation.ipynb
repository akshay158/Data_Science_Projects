```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
df = pd.read_csv('/Customer_data - customer_data.csv')

#Dataset overview
print(df.info())
print('-'*90)
print("\nFirst few rows of the dataset:")
display(df.head(10))
print('-'*90)

#Check for missing values
print("\nMissing values in the dataset:")
print(df.isnull().sum())
print('-'*90)

#Check for number of duplicate rows
print("\nNumber of duplicate rows in the dataset:")
print(df.duplicated().sum())
```

```python
#Only column Total Charges has no value in that cells that has 0 tenure. Hence we can keep the value of cells to 0
df.loc[df['tenure']==0, 'TotalCharges'] = 0

#Replacing 'No phone service' to 'No' in MultipleLines column
df['MultipleLines'] = df['MultipleLines'].replace('No phone service', 'No')
print(f"Unique values in MultipleLines column after replacement is: {df['MultipleLines'].unique()}")

#Replacing 'No internet service' to 'No' in Multiple columns
column_to_replace = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
for column in column_to_replace:
    df[column] = df[column].replace('No internet service', 'No')
    print(f"Unique values in {column} column after replacement is: {df[column].unique()}")

```

```python
# Converting categorical columns to numeric(one-hot encoding)
df = pd.get_dummies(df,drop_first=True)

# Feature scaling (standardization)
scaler = StandardScaler()
df[['tenure', 'MonthlyCharges', 'TotalCharges']] = scaler.fit_transform(df[['tenure', 'MonthlyCharges', 'TotalCharges']])

#Saving the process file
processed_dataset = 'processed_Customer_dataset.csv'
df.to_csv(processed_dataset, index=False)

print(f"\nProcessed dataset saved to: {processed_dataset}")
```
