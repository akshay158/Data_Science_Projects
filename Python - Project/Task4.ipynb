```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
ae = pd.read_csv("FEV-data-Excel.xlsx - Auta elektryczne.csv")

# Select only numerical columns from the dataset
numerical_ae = ae.select_dtypes(include=[float, int])

# Calculate the correlation matrix
correlation_matrix = pd.DataFrame(numerical_ae.corr())

# Display the correlation matrix
print("Correlation matrix:")
display(correlation_matrix)
```

```python

# Plot a heatmap to visualize the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Columns")
plt.show()
```

```python
# As per Heatmap one of the correlation that has the highest positive value is Engine power [KM] and Maximum torque [Nm]
# Let's see how this relation looks as per scatter plot

plt.figure(figsize=(8, 6))
sns.barplot(data=ae, x='Engine power [KM]', y='Maximum torque [Nm]')
plt.title("Engine Power vs. Maximum torque")
plt.xlabel("Engine Power [KM]")
plt.ylabel("Maximum torque [Nm]")
plt.show()

# As it can be seen in the graph, more the power of the Engine, more is the Gross price of the vehicle.
```

```python
# As per Heatmap one of the other correlation that has the highest negative value is Maximum speed [kph] and Acceleration 0-100 kph [s]
# Let's see how this relation looks as per scatter plot

plt.figure(figsize=(8, 6))
sns.barplot(data=ae, y='Acceleration 0-100 kph [s]', x= 'Maximum speed [kph]')
plt.title("Maximum speed [kph] vs. Acceleration")
plt.xlabel("Maximum speed [kph]")
plt.ylabel("Acceleration 0-100 kph [s]")
plt.show()

#As per the graph, it can be concluded that Engine Power has negative correlation with Acceleration of vehicle from 0-100 kph which means
# more the Engine Power, less is the Acceleration of the Vehicle
```
